```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
import yfinance as yf
tickerSymbol = 'ICICIBANK.NS'
tickerData = yf.Ticker(tickerSymbol)
ICICIBANK = tickerData.history(period='1d', start='2010-03-1')
print("ICICIBANK shares " ,"\n",ICICIBANK.tail())

tickerSymbol = 'SBIN.NS'
tickerData = yf.Ticker(tickerSymbol)
SBIBANK= tickerData.history(period='1d', start='2010-03-1')
print("\n","SBIBANK shares ","\n" ,SBIBANK.tail())
```

```
ICICIBANK = ICICIBANK.rename(columns={'Close' : 'ICICI_Shares'})
SBIBANK  = SBIBANK.rename(columns={'Close' : 'SBI_Shares'})
icici_sbi_data = pd.DataFrame(data = [ICICIBANK['ICICI_Shares'],SBIBANK['SBI_Shares']]).T
icici_sbi_data.plot(figsize=(10, 7))
plt.legend()
plt.title("Adjusted Close Price of Two companies ", fontsize=16)
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

```
ICICIBANK_returns = np.log(ICICIBANK['ICICI_Shares'] / ICICIBANK['ICICI_Shares'].shift(1))
ICICIBANK_returns.tail(20)
```

```
SBIBANK_returns = np.log(SBIBANK['SBI_Shares'] / SBIBANK['SBI_Shares'].shift(1))
SBIBANK_returns.tail(20)
```

```
"""##ICICI"""

print ("ICICI returns mean = ", ICICIBANK_returns.mean(),"\n\n",
       "ICICI returns mean * 250 = ",ICICIBANK_returns.mean()*250,"\n\n",
       "ICICI returns Standard Deviation = ",ICICIBANK_returns.std(),"\n\n",
       "ICICI returns Standard Deviation * 250 * 0.5 = ",ICICIBANK_returns.std()*250**0.5)
```

```
"""## SBI"""

print ("SBI returns mean = ", SBIBANK_returns.mean(),"\n\n",
       "SBI returns mean * 250 = ",SBIBANK_returns.mean()*250,"\n\n",
       "SBI returns Standard Deviation = ",SBIBANK_returns.std(),"\n\n",
       "SBI returns Standard Deviation * 250 * 0.5 = ",SBIBANK_returns.std()*250**0.5)
```

```
"""## Final Results"""

print(ICICIBANK_returns.mean()*250)
print(SBIBANK_returns.mean()*250)
```

```
icici_sbi_returns = pd.DataFrame(data = [ICICIBANK_returns,SBIBANK_returns]).T
print ("icici vs sbi mean * 250","\n\n",icici_sbi_returns[['ICICI_Shares','SBI_Shares']].mean()*250,"\n\n\n\n",
"icici vs sbi mean * 250 square 0.5 ","\n\n",icici_sbi_returns[['ICICI_Shares','SBI_Shares']].mean()*250**0.5)
```

```
icici_sbi_returns.plot(figsize=(10, 7))
plt.legend()
plt.title("Returns of Two companies ", fontsize=16)
plt.ylabel('Returns', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
